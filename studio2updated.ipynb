```python
import requests
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
import time
import numpy as np
from nltk.tokenize.casual import TweetTokenizer
import re
import pandas as pd
nltk.download('vader_lexicon')
from emot import emo_unicode
```

```python
def convert_emojis(text):
    for emot in emo_unicode.UNICODE_EMO:
        if emot in text:
            text = re.sub(r'('+emot+')', "_".join(emo_unicode.UNICODE_EMO[emot].replace(",","").replace(":","").split())+" ", text)
            text = text.replace("_"," ")
            text = text.replace("-"," ")
    return text
```

```python
def analyze_user(user):
    a = np.random.rand()
    url = 'https://www.instagram.com/' + user
    r = requests.get(url).text

    start = '"edge_followed_by":{"count":'
    end = '},"followed_by_viewer"'
    
    followers = (r[r.find(start)+len(start):r.rfind(end)])
    print("followers",followers)
    start = '"edge_follow":{"count":'
    end = '},"follows_viewer"'
    
    follows = r[r.find(start)+len(start):r.rfind(end)]
    print("follows",follows)
    
    bio_start = '"user":{"biography":'
    bio_end = ',"blocked_by_viewer":'
    
    bio = (r[r.find(bio_start)+len(bio_start):r.rfind(bio_end)])
    
#     print(bio)
    
    print(min(10*(int(follows)/int(followers)-0.5),5))
    return [user,follows,followers,min(10*(int(follows)/int(followers)-0.5),5),bio]
```

```python
def analyze_brand(brand_par):
    
    brand_file = brand_par+".csv"
    raw_data = pd.read_csv('parade.csv')
    list_new = raw_data.values.tolist()
    
    sid = SentimentIntensityAnalyzer()

    user_list = []

    for comment in list_new:
        a = np.random.rand()
    #   print(comment[1])
        text_comment = convert_emojis(comment[1])
        print(text_comment)
        scores_dict = sid.polarity_scores(comment[1])
        scores_list = list(scores_dict.items())
        scores_array = np.asarray(scores_list)

        ''' WARNING: Note that we multiply by -1 to get the highest score first from argsort '''
        sorted_array = np.argsort(-1*scores_array[[0,2],1].astype(float))

    #     print("sorted_array", sorted_array)
        max_ind = sorted_array[0]
        print(max_ind)
        # negative user_score assigned to influencer, positive to non-influencer (range is -5 to 5)
        user_score = analyze_user(comment[0])
        print("user_score",user_score)

        # multiplies sentiment by user score. non-influencer with positive sentiment adds points and viceversa
        mult_factor = float(user_score[3])
        print("mult_factor",mult_factor)

        comment_score = scores_array[[0,2]][max_ind][1].astype(float) * (-1)**((max_ind)+1)
        print("comment_score",comment_score)
        comment_score = np.multiply(comment_score , mult_factor)
        print("comment_score",comment_score)

        comment_data = [comment[0],text_comment,comment_score]
        print(comment_data)
        user_list.append(comment_data)
        
        ''' WARNING: Reactivate Line Below in Production to avoid Rate Limit on Instagram'''
        time.sleep(60+120*(a))

    with open ("users_data_"+brand_par+".txt", 'w') as file:

        for i in user_list:
            str_i = [str(j) for j in i]
            str_i = ";;;".join(str_i)
            file.write(str_i+"\n")
            file.close
```

```python
brand_list = ["revolve(1).csv","parade.csv"]

for brand in brand_list:
    analyze_brand(brand)
```
