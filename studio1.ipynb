```python
filename = 'revolve.txt'
file = open(filename, 'rt')
text = file.read()
file.close()
```

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
#print(tokens)
```

```python
tokens = [w.lower() for w in tokens]
```

```python
import string
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in tokens]
```

```python
words = [word for word in stripped if word.isalpha()]
```

```python
from nltk.corpus import stopwords
#stop_words = stopwords.words('english')
stop_words = [i for i in stopwords.words('english')]
```

```python
my_stop_word = ['owned','us','million','information','brand','brands','customer','customers','consumer','consumers','fashion','marketing','revolve']
stop_words.extend(my_stop_word)
#stop_words = set(stop_words)
```

```python
words = [w for w in words if w not in stop_words]
```

```python
print(words[:160])
```

```python
word_list = ' '.join(words)
```

```python
#print (word_list)
```

```python
def generate_wordcloud(text): # optionally add: stopwords=STOPWORDS and change the arg below
    wordcloud = WordCloud(font_path='/Library/Fonts/Verdana.ttf',
                          relative_scaling = 1.0,
                          stopwords = stop_words # set or space-separated string
                          ).generate(text)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
```

```python
with open('mynewfile.txt', 'w') as f:
    for item in word_list:
        f.write(item)
file_name = 'mynewfile.txt'
file_new= open(file_name, 'rt')
final_text = file_new.read()
generate_wordcloud(final_text)
```
